```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
x
```

```python
y
```

```python

import matplotlib.pyplot as plt
%matplotlib inline

```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_xlabel("x")
ax.set_ylabel('Y')
ax.set_title('title')
ax.plot(x,y)
```

```python
fig =plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot()
ax2.plot()
```

```python
fig =plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
#insert  figure in ax1 
ax1.set_xlabel('X')
ax1.plot(x,y,'r')
# insert figure in ax2
ax2.set_label('Y')
ax2.plot(x,y,'r')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax1.plot()
ax2.plot()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
# insert figure in ax1
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.plot(x,z)
#  insert figure in ax2
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.plot(x,y)
```
